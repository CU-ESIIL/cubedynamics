## 2. Build a demo event

```
import cubedynamics as ccm
event = ccm.make_demo_event(n_days=6, random_state=2)
event[['event_day', 'date']].head()
```

The helper returns a GeoDataFrame in EPSG:5070.  Each row represents one day of growth.

## 3. Plot the ruled time hull

```
fig, ax = ccm.plot_ruled_time_hull(event, n_theta=96, smooth_over_z=3)
fig
```

The walls are assembled from equal-angle support functions so the 3D shape is smooth
and easy to interpret.
