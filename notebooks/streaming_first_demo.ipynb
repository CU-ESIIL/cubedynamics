# Streaming-first cube rendering
Preview how the viewer streams slices and shows progress.

```
from cubedynamics import pipe, verbs as v
import xarray as xr
import numpy as np
```

```
big = xr.DataArray(np.random.rand(40, 40, 40), dims=('time','y','x'), name='ndvi')
pipe(big) | v.plot(title='Streaming preview', thin_time_factor=5, show_progress=True)
```
