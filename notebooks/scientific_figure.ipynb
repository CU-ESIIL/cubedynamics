# Scientific-quality cube figure
Caption, legend, and studio theme.

```
from cubedynamics.plotting.cube_plot import CubePlot
import xarray as xr
import numpy as np
```

```
cube = xr.DataArray(np.random.rand(6, 12, 12), dims=('time','y','x'), name='ndvi')
```

```
figure = (CubePlot(cube, title='Fire season NDVI')
          .aes(fill='ndvi')
          .stat_time_anomaly(time_dim='time')
          .geom_cube()
          .scale_fill_continuous(center=0, palette='diverging', name='NDVI z-score')
          .coord_cube(elev=28, azim=55)
          .theme_cube_studio()
          .annot_plane(axis='time', value=3, text='Smoke plume'))
figure.caption = {'id': 4, 'title': 'Fire season cube', 'text': 'Markdown + $\nabla$ ready.'}
figure
```
