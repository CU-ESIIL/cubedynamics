# Cube Grammar Tutorial
This notebook demonstrates the CubePlot grammar-of-graphics components.

```
from cubedynamics import pipe, verbs as v
from cubedynamics.plotting.cube_plot import CubePlot
import xarray as xr
import numpy as np
```

```
cube = xr.DataArray(np.random.rand(6, 10, 10), dims=('time','y','x'), name='ndvi')
```

```
p = (CubePlot(cube, title='NDVI anomaly')
     .aes(fill='ndvi')
     .stat_time_anomaly(time_dim='time')
     .geom_cube()
     .scale_fill_continuous(center=0, palette='diverging')
     .coord_cube(elev=30, azim=45)
     .theme_cube_studio()
     .annot_plane(axis='time', value=2, text='Fire event'))
p
```
