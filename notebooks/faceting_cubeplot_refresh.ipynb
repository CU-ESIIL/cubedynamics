# Faceting cubes
Comparing scenarios with facet_wrap and facet_grid.

```
from cubedynamics.plotting.cube_plot import CubePlot
import xarray as xr
import numpy as np
```

```
data = xr.DataArray(np.random.rand(4,3,8,8), dims=('scenario','time','y','x'), name='ndvi', coords={'scenario':['base','wetter','drier','fire']})
```

```
fw = (CubePlot(data)
      .aes(fill='ndvi')
      .facet_wrap(by='scenario', ncol=2)
      .geom_cube()
      .scale_fill_continuous(palette='sequential')
      .theme_cube_studio())
fw
```

```
fg = (CubePlot(data.sel(scenario=['base','wetter']))
      .facet_grid(row='scenario', col=None)
      .geom_cube()
      .scale_fill_continuous(center=0)
      .annot_plane(axis='time', value=1, text='El Ni√±o'))
fg
```
